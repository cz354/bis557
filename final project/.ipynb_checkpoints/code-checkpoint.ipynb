## Data Processing

```python
train_features=pd.read_csv("./lish-moa/train_features.csv")
test_features = pd.read_csv("./lish-moa/test_features.csv")
train_targets_scored = pd.read_csv("./lish-moa/train_targets_scored.csv")
train_targets_nonscored = pd.read_csv("./lish-moa/train_targets_nonscored.csv")
train_drug = pd.read_csv("./lish-moa/train_drug.csv")
```

```python
train_features1=train_features.copy()
test_features1=test_features.copy()

train_features1.loc[:, 'cp_time'] = train_features['cp_time'].map({24: 0, 48: 1, 72: 2})
train_features1.loc[:, 'cp_type'] = train_features['cp_type'].map({'trt_cp': 0, 'ctl_vehicle': 1})
train_features1.loc[:, 'cp_dose'] = train_features['cp_dose'].map({'D1':0, 'D2':1})

test_features1.loc[:, 'cp_time'] = test_features['cp_time'].map({24: 0, 48: 1, 72: 2})
test_features1.loc[:, 'cp_type'] = test_features['cp_type'].map({'trt_cp': 0, 'ctl_vehicle': 1})
test_features1.loc[:, 'cp_dose'] = test_features['cp_dose'].map({'D1':0, 'D2':1})
```

## Feature selection

### Dimensionality Reduction (PCA) https://www.mikulskibartosz.name/pca-how-to-choose-the-number-of-components/


```python
for col in (gene+cell):
    transformer = QuantileTransformer(random_state=129, output_distribution="normal")
    train1 = train_features[col].values.reshape(len(train_features[col].values), 1)
    transformer.fit(train1)

    train_features[col] = transformer.transform(train1).reshape(1, len(train_features[col].values))[0]
    test2=test_features[col].values.reshape(len(test_features[col].values), 1)
    test_features[col] = transformer.transform(test2).reshape(1, len(test_features[col].values))[0]
    
data_rescaled_gene = pd.concat([pd.DataFrame(train_features[gene]), pd.DataFrame(test_features[gene])])
data_rescaled_cell = pd.concat([pd.DataFrame(train_features[cell]), pd.DataFrame(test_features[cell])])
```

```python
pca = PCA().fit(data_rescaled_gene)

plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 773, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 750, step=50)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
pca = PCA(n_components = 0.95)
pca.fit(data_rescaled_gene)
reduced_gene = pca.transform(data_rescaled_gene)
gene_pca = pd.DataFrame(reduced_gene, columns=[f'PCA_g-{i}' for i in range(603)])
train_gene_pca=gene_pca.iloc[0:23814]
test_gene_pca=gene_pca.iloc[23814:].reset_index(drop=True)
```

```python
pca = PCA().fit(data_rescaled_cell)

plt.rcParams["figure.figsize"] = (12,6)

fig, ax = plt.subplots()
xi = np.arange(1, 101, step=1)
y = np.cumsum(pca.explained_variance_ratio_)

plt.ylim(0.0,1.1)
plt.plot(xi, y, marker='o', linestyle='--', color='b')

plt.xlabel('Number of Components')
plt.xticks(np.arange(0, 120, step=10)) #change from 0-based array index to 1-based human-readable label
plt.ylabel('Cumulative variance (%)')
plt.title('The number of components needed to explain variance')

plt.axhline(y=0.95, color='r', linestyle='-')
plt.text(0.5, 0.85, '95% cut-off threshold', color = 'red', fontsize=16)

ax.grid(axis='x')
plt.show()
```

```python
pca = PCA(n_components = 0.95)
pca.fit(data_rescaled_cell)
reduced_cell = pca.transform(data_rescaled_cell)
cell_pca = pd.DataFrame(reduced_cell, columns=[f'PCA_c-{i}' for i in range(84)])
train_cell_pca=cell_pca.iloc[0:23814]
test_cell_pca=cell_pca.iloc[23814:].reset_index(drop=True)
```

```python
train_features1 = pd.concat((train_features1, train_gene_pca), axis=1)
train_features1 = pd.concat((train_features1, train_cell_pca), axis=1)

test_features1 = pd.concat((test_features1, test_gene_pca), axis=1)
test_features1 = pd.concat((test_features1, test_cell_pca), axis=1)

pca_fea=['cp_type','cp_time','cp_dose',]+[f'PCA_c-{i}' for i in range(84)]+[f'PCA_g-{i}' for i in range(603)]
```

## Model

```python
target = [i for i in train_targets_scored.columns if i != 'sig_id']
X = train_features1[pca_fea].values 
y = train_targets_scored[target].values
X_test = test_features1[pca_fea].values
rm=X[:,0]==1
X_train = X[~rm,:]
y_train = y[~rm,:]
```

### XGboost

```python
kf = KFold(n_splits = 5)
val_preds = np.zeros(y_train.shape)
test_preds = np.zeros((X_test.shape[0], y_train.shape[1]))

params = {'colsample_bytree': 0.6,
          'gamma': 3.7,
          'learning_rate': 0.05,
          'max_delta_step': 2,
          'max_depth': 10,
          'min_child_weight': 31.6,
          'n_estimators': 166,
          'subsample': 0.9,
          'verbosity':0
         }

fit1 = MultiOutputClassifier(XGBClassifier(**params, tree_method='gpu_hist'))
```

```python
for (f, (id1, id2)) in enumerate(kf.split(X_train)):
    print(f)
    X_train_fold, X_val_fold = X_train[id1,:], X_train[id2,:]
    y_train_fold = y_train[id1,:]
    y_val_fold =y_train[id2,:]
    

    fit1.fit(X_train_fold,y_train_fold )
    
    # generate the prediction
    val_pred = fit2.predict_proba(X_val_fold)
    test_pred = fit2.predict_proba(X_test)
    
    for i in range(0,y_train.shape[1]):
        val_preds[id2,i] = val_pred[i][:,1]
        test_preds[:,i] += test_pred[i][:,1]/5 
```

```python
print("log-loss for XGBoost")
log_loss(np.ravel(y_train), np.ravel(val_preds))
```

### multiclass logistic regression

```python
kf = KFold(n_splits = 5)
val_preds = np.zeros(y_train.shape)
test_preds = np.zeros((X_test.shape[0], y_train.shape[1]))

fit2= MultiOutputClassifier(LogisticRegression(max_iter=1000,tol=0.1, C = 0.5))
```

```python
for (f, (id1, id2)) in enumerate(kf.split(X_train)):
    print(f)
    X_train_fold, X_val_fold = X_train[id1,:], X_train[id2,:]
    y_train_fold = y_train[id1,:]
    y_val_fold =y_train[id2,:]
    
    index=np.where(y_train_fold.sum(axis=0)==0)[0]
    if len(index)!=0:
        y_train_fold[0,index] = 1
    

    fit2.fit(X_train_fold,y_train_fold )
    
    # generate the prediction
    val_pred = fit2.predict_proba(X_val_fold)
    test_pred = fit2.predict_proba(X_test)
    
    for i in range(0,y_train.shape[1]):
        val_preds[id2,i] = val_pred[i][:,1]
        test_preds[:,i] += test_pred[i][:,1]/5      
    
```

```python
print("log-loss for multilabel logistic regression")
log_loss(np.ravel(y_train), np.ravel(val_preds))
```

### Multilabel neural network

```python
X_train=train_features1.loc[train_features1["cp_type"]==0]
y_train=train_targets_scored.loc[train_features1["cp_type"]==0]
X_train=X_train[pca_fea]
y_train=y_train[target]
X_test = test_features1[pca_fea]
```

```python
kf = KFold(n_splits = 5)
val_preds = np.zeros(y_train.shape)
test_preds = np.zeros((X_test.shape[0], y_train.shape[1]))


fit3 = tf.keras.Sequential([
    tf.keras.layers.Input(690),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.3),
    tfa.layers.WeightNormalization(tf.keras.layers.Dense(2048, activation="relu")),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    tfa.layers.WeightNormalization(tf.keras.layers.Dense(1048, activation="relu")),
    tf.keras.layers.BatchNormalization(),
    tf.keras.layers.Dropout(0.5),
    tfa.layers.WeightNormalization(tf.keras.layers.Dense(206, activation="sigmoid"))
    ])
fit3.compile(optimizer=tfa.optimizers.Lookahead(tf.optimizers.Adam(), sync_period=10),
                  loss='binary_crossentropy', )
```

```python
for (f, (id1, id2)) in enumerate(kf.split(X_train)):
        print(f)
        X_train_fold, X_val_fold = X_train.iloc[id1], X_train.iloc[id2]
        y_train_fold = y_train.iloc[id1]
        y_val_fold =y_train.iloc[id2]
        
        checkpoint_path = f'Fold:{f}.hdf5'
        reduce_lr_loss = ReduceLROnPlateau(monitor='val_loss', factor=0.1, patience=3, verbose=1, epsilon=1e-4, mode='min')
        cb_checkpt = ModelCheckpoint(checkpoint_path, monitor = 'val_loss', verbose = 0, save_best_only = True,
                                     save_weights_only = True, mode = 'min')
        
    
        fit3.fit(X_train_fold.values,
                  y_train_fold.values,
                  validation_data=( X_val_fold.values, y_val_fold.values),
                  epochs=35, batch_size=128,
                  callbacks=[reduce_lr_loss, cb_checkpt], verbose=2
                 )
        
        fit3.load_weights(checkpoint_path)
        val_pred = fit3.predict(X_val_fold.values)
        test_pred = fit3.predict(X_test.values)
        for i in range(0,y_train.shape[1]):
            val_preds[id2,i] = val_pred[:,i]
            test_preds[:,i] += test_pred[:,i]/5 
```

```python
print("log-loss for multilabel NN")
log_loss(np.ravel(y_train), np.ravel(val_preds))
```
